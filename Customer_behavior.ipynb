```python
import pandas as pd 

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
#Verificar Valores Vazios
df.isnull().sum()
```

```python
#substituindo valores vazios pela media dos valores
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
#Verificando novamente os valores nulos e percebe-se que preechemos todos
df.isnull().sum()
```

```python
#Alterar os nomes das columas para facilitar o trabalho mais a frente 
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
#Verificando se o nome foi alterado 
df.columns
```

```python
#Criar uma coluna com faixa etária 
labels = ['Young Adult', 'Adult', 'Middle-age', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
#criar a coluna purchase_frequency_days
frequency_mapping = {
    'Fortnightly':14,
    'Weekly': 7,
    'Monthly':30,
    'Quartely': 90,
    'Bi-Weekly':14,
    'Annualy': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```

```python
df[['promo_code_used', 'discount_applied']].head(10)
```

```python
#Verificando se as tabelas são praticamente a mesma coisa, se for iremos excluir uma
(df['promo_code_used'] == df['discount_applied']).all()
```

```python
#são iguais, vamos remover o promo_code_used
df = df.drop('promo_code_used', axis=1)
```
